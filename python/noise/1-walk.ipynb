```python
def main():
    '''Create example chart'''
    
    x = np.array(range(len(data))) / 5
    y = data

    fig = plt.figure(figsize=(FIG_WIDTH, FIG_HEIGHT))
    
    plt.scatter(x, y, label ='Unfiltered data')
    
    plt.title('Walking data captured at 5 Hz - u-blox M10 chipset')

    plt.xlabel("Time (s)")
    plt.ylabel("Speed (km/h)")
    
    ax = fig.get_axes()[0]

    comment = 'Still arms = {:.1f} to {:.1f} km/h\nmean = {:.1f} km/h\n\nSwinging arms = {:.1f} to {:.1f} km/h\nmean = {:.1f} km/h'.format(
        np.min(data[:100]), np.max(data[:100]), np.mean(data[:100]), np.min(data[110:]), np.max(data[110:]), np.mean(data[110:]))
    fig.text(0.16, 0.18, comment)

    plt.show()

    fname = os.path.join(IMG_FOLDER, '1-walk.png')
    fig.savefig(fname, bbox_inches='tight')


if __name__ == '__main__':
    projdir = os.path.realpath(os.path.join(sys.path[0], "..", ".."))

    main()
    
    print(os.linesep + 'All done!')
```
